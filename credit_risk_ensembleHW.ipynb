## Split the Data into Training and Testing

```python
# Create our features
X = df.drop(columns="loan_status") # YOUR CODE HERE
X = pd.get_dummies(X)


# Create our target
y = df["loan_status"] # YOUR CODE HERE
```

```python
X.describe()
```

```python
# Check the balance of our target values
# YOUR CODE HERE
y.value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
# YOUR CODE HERE
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
from sklearn.preprocessing import StandardScaler
# YOUR CODE HERE
sc = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
# YOUR CODE HERE
X_scaler = sc.fit(X_train)
```

```python
# Scale the training and testing data
# YOUR CODE HERE
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
# YOUR CODE HERE
balanced_rf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
balanced_rf.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
# YOUR CODE HERE
y_pred_bal_rf = balanced_rf.predict(X_test_scaled)
balanced_accuracy_score(y_test, y_pred_bal_rf)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
# YOUR CODE HERE
confusion_matrix(y_test, y_pred_bal_rf)
```

```python
# Print the imbalanced classification report
# YOUR CODE HERE
print(classification_report_imbalanced(y_test, y_pred_bal_rf))
```

### Easy Ensemble Classifier

```python
# Train the Classifier
from imblearn.ensemble import EasyEnsembleClassifier
# YOUR CODE HERE
ez = EasyEnsembleClassifier(n_estimators=100,random_state=1)
ez.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
# YOUR CODE HERE
y_pred_ez = ez.predict(X_test_scaled) 
balanced_accuracy_score(y_test, y_pred_ez)
```

```python
# Display the confusion matrix
# YOUR CODE HERE
confusion_matrix(y_test, y_pred_ez)
```

```python
# Print the imbalanced classification report
# YOUR CODE HERE
print(classification_report_imbalanced(y_test, y_pred_ez))
```
